### **Looping through Arrays in a superior Fashion**

---



```python
g = np.arange(2,21,2).reshape(5,2)
g
```

```python
### Normal Loop through elements, as Row by Row
for x in np.nditer(g, order = 'c'):
  print(x)
```

```python
### loop through Columns to get elements
for c in np.nditer(g, order ='f'):
  print(c)
```

```python
### Loop to get Entire Columns

for v in np.nditer(g, order ='f', flags = ['external_loop']):
  print(v)
```

### **Looping through 2 objects at the same timeimage.png**

---



```python
h = np.arange(1,20,2).reshape((1,10))
j = np.arange(2,22,2).reshape((10,1))
print(h)
print(j)
```

```python
for n,m in np.nditer([h,j]):
  print(n,m)
```
